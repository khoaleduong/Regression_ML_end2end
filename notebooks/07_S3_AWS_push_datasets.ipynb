```python
import boto3
from pathlib import Path
```

```python
# --- Configuration ---
# Replace with your actual S3 bucket and region details
bucket_name = "your-bucket-name"
region_name = "your-region-name"

# --- Directory Setup ---
# Define paths relative to the project root for better portability
project_root = Path.cwd().parent
local_data_dir = project_root / "data" / "processed"
local_model_dir = project_root / "models"

# Initialize the S3 client using boto3
s3_client = boto3.client("s3", region_name=region_name)

def upload_file_to_s3(local_file_path, s3_key):
    if not local_file_path.exists():
        print(f"File {local_file_path} does not exist.")
        return
    print(f"Uploading {local_file_path} to S3 bucket {bucket_name}...")
    s3_client.upload_file(str(local_file_path), bucket_name, s3_key)
    print(f"Uploaded {local_file_path} to S3 bucket {bucket_name}.")
```

```python
# upload datasets
for data_file in local_data_dir.glob("*.csv"):
    s3_key = f"data/processed/{data_file.name}"
    upload_file_to_s3(data_file, s3_key)
# upload models
upload_file_to_s3(local_model_dir / "regression.pkl", "models/regression.pkl")
```
