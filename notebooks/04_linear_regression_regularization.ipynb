```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
```

```python
train_df = pd.read_csv("/Users/leduongkhoa/Regression_MachineLearning_End2End/data/processed/encoded_train.csv")
eval_df = pd.read_csv("/Users/leduongkhoa/Regression_MachineLearning_End2End/data/processed/encoded_eval.csv")
```

```python
target_col = 'price'
X_train = train_df.drop(columns=[target_col])
y_train = train_df[target_col]

X_eval = eval_df.drop(columns=[target_col])
y_eval = eval_df[target_col]
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_eval_scaled = scaler.transform(X_eval)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred = lr.predict(X_eval_scaled)

print("Linear Regression Performance:")
print(f"Mean absolute Error (MAE): {mean_absolute_error(y_eval, y_pred)}")
print(f"Root Mean Squared Error (RMSE): {root_mean_squared_error(y_eval, y_pred)}")
print(f"R^2 Score: {r2_score(y_eval, y_pred)}")
```

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)
y_pred_ridge = ridge.predict(X_eval_scaled)
print("\nRidge Regression Performance:")
print(f"Mean absolute Error (MAE): {mean_absolute_error(y_eval, y_pred_ridge)}")
print(f"Root Mean Squared Error (RMSE): {root_mean_squared_error(y_eval, y_pred_ridge)}")
print(f"R^2 Score: {r2_score(y_eval, y_pred_ridge)}")
```

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train_scaled, y_train)
y_pred_lasso = lasso.predict(X_eval_scaled)

print("\nLasso Regression Performance:")
print(f"Mean absolute Error (MAE): {mean_absolute_error(y_eval, y_pred_lasso)}")
print(f"Root Mean Squared Error (RMSE): {root_mean_squared_error(y_eval, y_pred_lasso)}")
print(f"R^2 Score: {r2_score(y_eval, y_pred_lasso)}")
```

```python
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train_scaled, y_train)
y_pred_elastic = elastic.predict(X_eval_scaled)

print("\nElasticNet Regression Performance:")
print(f"Mean absolute Error (MAE): {mean_absolute_error(y_eval, y_pred_elastic)}")
print(f"Root Mean Squared Error (RMSE): {root_mean_squared_error(y_eval, y_pred_elastic)}")
print(f"R^2 Score: {r2_score(y_eval, y_pred_elastic)}")
```
